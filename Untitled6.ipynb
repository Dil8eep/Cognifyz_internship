## **Task 3**
1)Plot the locations of restaurants on a
map using longitude and latitude
coordinates.

2)Identify any patterns or clusters of
restaurants in specific areas.

```python
# Plot the locations of restaurants on a map using longitude and latitude coordinates
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(10, 8))
plt.scatter(data['Longitude'], data['Latitude'], marker='o', c='blue', s=100)


# Set plot title and axis labels
plt.title('Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Identify any patterns or clusters of restaurants in specific areas.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
locations = data[['Latitude', 'Longitude']]
scaler = StandardScaler()
scaled_locations = scaler.fit_transform(locations)
inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_locations)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()

optimal_k = 3  # Replace with the value you determined from the Elbow Method
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['cluster'] = kmeans.fit_predict(scaled_locations)

# Create a scatter plot of restaurant locations with different colors for clusters
plt.figure(figsize=(10, 8))
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

for i in range(optimal_k):
    cluster_df = data[data['cluster'] == i]
    plt.scatter(cluster_df['Longitude'], cluster_df['Latitude'], c=colors[i], label=f'Cluster {i}')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Clusters')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data.columns
```

```python
# Analyze the text reviews to identify the most common positive and negative keywords.
```

```python
reviews = data['Rating text']

# Join all reviews into one long text
all_reviews = ' '.join(reviews)

# Tokenize the text (split it into words)
words = all_reviews.split()

word_freq = pd.Series(words).value_counts()

# Define stopwords (words to exclude)
stopwords = ['the', 'and', 'in', 'to', 'of', 'a', 'for', 'with', 'on', 'this']

word_freq = word_freq[~word_freq.index.isin(stopwords)]

top_words = word_freq.head(20)
plt.figure(figsize=(12, 6))
top_words.plot(kind='bar', color='green')
plt.title('Most Common Positive Keywords')
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Plot the most common negative keywords
bottom_words = word_freq.tail(20)

plt.figure(figsize=(12, 6))
bottom_words.plot(kind='bar', color='red')
plt.title('Most Common Negative Keywords')
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the average length of reviews and explore if there is a relationship between review length and rating.
import pandas as pd

# Load your dataset into a Pandas DataFrame
# df = pd.read_csv('your_dataset.csv')

# Assuming 'Reviews' and 'Ratings' are the relevant columns
reviews = data['Rating text']
ratings = data['Aggregate rating']

# Calculate the length of each review (number of words)
review_lengths = reviews.apply(lambda x: len(str(x).split()))

# Add the review lengths as a new column in the DataFrame
data['Rating text'] = review_lengths

# Calculate the average review length
average_review_length = review_lengths.mean()
print("Average Review Length (in words):", average_review_length)

# Explore the relationship between review length and rating
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(data=['Rating text'],ratings,alpha=0.5)
plt.title('Relationship between Review Length and Rating')
plt.xlabel('Review Length (in words)')
plt.ylabel('Rating')
plt.show()
```
